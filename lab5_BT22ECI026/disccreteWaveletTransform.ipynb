```python
import cv2
import pywt
import matplotlib.pyplot as plt
```

```python
# Read the image
i = cv2.imread('../dip_img.jpeg')

# Convert to grayscale
i_gray = cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)

# Apply 2D Discrete Wavelet Transform
LL, (LH, HL, HH) = pywt.dwt2(i_gray, 'haar')
```

```python
# Plot the four sub-bands
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.imshow(LL, cmap='gray')
plt.title('Approximation')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(LH, cmap='gray')
plt.title('Horizontal detail')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(HL, cmap='gray')
plt.title('Vertical detail')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(HH, cmap='gray')
plt.title('Diagonal detail')
plt.axis('off')
```

```python
# Sum of all components
x = LL + LH + HL + HH
```

```python
# Plot original and decomposed images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(i_gray, cmap='gray')
plt.title('Original image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(x, cmap='gray')
plt.title('Decomposed image')
plt.axis('off')

plt.show()
```
